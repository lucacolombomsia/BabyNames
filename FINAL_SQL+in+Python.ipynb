```python
import pandas as pd
import numpy as np
import pymysql

cnx = pymysql.connect(user='root', \
      password='luca', \
      host= '127.0.0.1', \
      port=3306, \
      db='world', \
      autocommit=True)

df_country = pd.read_sql_query('select * from Country', con=cnx)
df_city = pd.read_sql_query('select * from city', con=cnx)
df_lang = pd.read_sql_query('select * from countrylanguage', con=cnx)
```

## Question 1

```python
df_country[df_country.Population >  50000000].sort_values('Population', ascending=False).iloc[:10]
```

## Question 2

```python
df_country[df_country.Population > 0][['Name','Continent','Population']].groupby(
    'Continent', as_index=False).agg({'Name':'count','Population' : 'sum'}).sort_values(
    'Continent').rename(columns = {'Name':'Number_Countries'})
```

## Question 3

```python
df_city.merge(df_country[df_country.Code == 'USA'][['Code']], left_on = 'CountryCode', right_on = 'Code')\
[['Name','Population']].sort_values('Population', ascending = False).iloc[:10]
```

Note that the same question can be solved in a slightly easier way, since the join performed in the question is not really needed, as all the columns that are selected are already present in the "city" table. Below you can find the alternative solution.

```python
df_city[df_city.CountryCode == 'USA'][["Name", "Population"]].sort_values(
    "Population",ascending = False).iloc[:10].reset_index(drop = True)
```

## Question 4

```python
l=df_lang[df_lang.IsOfficial == 'T'].merge(df_country[['Code','Name','Population']],left_on = 'CountryCode',right_on='Code')\
[['Name','Language','Population','Percentage']]
l['Speakers'] = l.Percentage * l.Population /100
l[['Name','Language','Speakers']].sort_values('Speakers', ascending = False).iloc[:10]
```

## Question 5

```python
s = df_lang.merge(df_country[['Code','Name','Population']],left_on = 'CountryCode',right_on='Code')
s['Speakers'] = s.Population * s.Percentage / 100
s[['Language','Speakers']].groupby('Language',as_index=False).sum().sort_values('Speakers', ascending = False).iloc[:5]
```
