```python
sql = """select *
from country
where population > 50000000
order by population DESC limit 10"""
df_largepop = pd.read_sql_query(sql, con=cnx)
df_largepop
```

### Question 2

```python
sql = """select Continent, count(*) As Number_Countries, sum(population) As Population
from country
where population > 0
group by Continent
order by 1 ASC"""
df_continents = pd.read_sql_query(sql, con=cnx)
df_continents
```

```python
df_continents["ID"] = range(len(df_continents))
temp = df_continents[["Continent", "ID"]]

df = country[country["Population"] > 0][["Continent", "Population"]].groupby("Continent").agg(["count", "sum"])
df.columns = ['Number_Countries', 'Population']
df["Continent"] = df.index
result = pd.merge(df, temp, left_on='Continent', right_on='Continent' , how='inner', sort=False)
result = result.sort_values(by = "ID").drop(["ID"], axis = 1)
df_continents_pd = result[["Continent", "Number_Countries", "Population"]]
df_continents_pd
```

```python
sql = """select city.Name As City, city.population
from city
inner join country ON city.CountryCode = country.code
where country.code = 'USA'
order by city.population DESC limit 10"""
df_citypop = pd.read_sql_query(sql, con=cnx)
df_citypop
```

```python
df_citypop_pd = city[city["CountryCode"] == 'USA'][["Name", "Population"]].sort_values(by="Population", ascending = False).head(10)
df_citypop_pd.columns = ["City", "population"]
df_citypop_pd
```

### Question 4

```python
sql = """select country.Name, Language, (Percentage * population) / 100
from countrylanguage
inner join country on countrylanguage.CountryCode = country.code
where IsOfficial = True
order by 3 DESC limit 10"""
df_officiallang = pd.read_sql_query(sql, con=cnx)
df_officiallang
```

```python
left = countrylang[countrylang["IsOfficial"] == "T"].set_index("CountryCode")
df = country.join(left, on="Code", how="inner", lsuffix='_left', rsuffix='_right')[["Name", "Language", "Population", "Percentage"]]
df["(Percentage * population) / 100"] = df["Population"]*df["Percentage"]/100.0
df_officiallang_pd = df.drop(["Population", "Percentage"], axis = 1).sort_values(by="(Percentage * population) / 100", ascending = False).head(10)
df_officiallang_pd
```

### Question 5

```python
sql = """select Language, sum((Percentage * population) / 100)
from countrylanguage
inner join country ON countrylanguage.CountryCode = country.code
group by Language
order by 2 desc limit 5"""
df_lang = pd.read_sql_query(sql, con=cnx)
df_lang
```

```python
left = countrylang.set_index("CountryCode")
df = country.join(left, on="Code", how="inner", lsuffix='_left', rsuffix='_right')[["Language", "Population", "Percentage"]]
df["(Percentage * population) / 100"] = df["Population"]*df["Percentage"]/100.0
df_lang_pd = df.drop(["Population", "Percentage"], axis = 1).groupby("Language", as_index=False).sum().sort_values(by="(Percentage * population) / 100", ascending = False).head(5)
df_lang_pd
```
